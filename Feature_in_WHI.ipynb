```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_2015 = pd.read_csv('/content/2015.csv')
df_2016 = pd.read_csv('/content/2016.csv')
df_2017 = pd.read_csv('/content/2017.csv')
df_2018 = pd.read_csv('/content/2018.csv')
df_2019 = pd.read_csv('/content/2019.csv')
df_2020 = pd.read_csv('/content/2020.csv')
df_2021 = pd.read_csv('/content/2021.csv')
```

```python
df_2015.corr()
```

```python
#Create Feature variable X and Target variable y
X_2015 = df_2015.drop(['Happiness Score','Region','Country', 'Happiness Rank'], axis = 1)
y_2015 = df_2015['Happiness Score']
from sklearn.ensemble import RandomForestRegressor
#Feature selection by Random forest regressor
clf=RandomForestRegressor()
clf.fit(X_2015,y_2015)
```

```python
ranked_features=pd.Series(clf.feature_importances_,index=X_2015.columns)
ranked_features.nlargest(5).plot(kind='barh')
plt.show()
```
